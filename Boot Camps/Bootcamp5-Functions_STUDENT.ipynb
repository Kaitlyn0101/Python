## Task

Write a function that takes in an array of x values and returns the equation a*x^3 + b*x^2 + c*x + d

Write a second function that take in x and y values and plots the results.

Create a plot that contains at least 3 different lines using your two functions.

```python
import numpy as np
import matplotlib.pyplot as plt


def second_function(x_array, y_array, ls= None,c='b'):
    plt.plot(x_array, y_array,color=c, linestyle=ls)
    
def first_function(a, b, c, d, x_array):
    return a*x_array**3+b*x_array**2+ c*x_array+ d
    

    
y_array=first_function(1,2,3,4,x_array)
second_function(x_array, y_array)

y_array=first_function(2,4,6,8,x_array)
second_function(x_array, y_array,'--','r')

y_array=first_function(1,3,5,7,x_array)
second_function(x_array, y_array,':','g')
```
