__NESTED STATEMENTS__

```python
temperature = [0.0, 1.1, 3.0, 5.0, 7.5, -0.5, 0.4, 7.0, 4.3, 1.2] # temperature in degrees celsius
relative_humidity = [40, 50, 60, 70, 80, 90] #relative humidity list (optional to use)

forecast=[]
for i in range(len(temperature)):             #create a nested for loop in order to loop over both lists
        for h in range(len(relative_humidity)):
            if relative_humidity[h]>=40 and relative_humidity[h]<50 and temperature[i]<=5.0:      #humidity range 40-50
                 forecast.append("snow")
            elif relative_humidity[h] >=50 and relative_humidity[h]<60 and temperature[i]<=4.0:   #humidity range 50-60
                  forecast.append("snow")
            elif relative_humidity[h] >=60 and relative_humidity[h]<70 and temperature[i]<=3.1:   #humidity range 60-70
                 forecast.append("snow")
            elif relative_humidity[h] >=70 and relative_humidity[h]<80 and temperature[i]<=2.5:   #humidity range 70-80
                 forecast.append("snow")
            elif relative_humidity[h] >=80 and relative_humidity[h]<90 and temperature[i]<=2.0:   #humidity range 80-90
                 forecast.append("snow")
            elif relative_humidity[h] >=90 and temperature[i]<=1.0:                               #humidity range 90-100
                forecast.append("snow")
            else:
                 forecast.append("rain")
                    
print(forecast)
```

__FITTING FUNCTIONS__

```python
from scipy.optimize import curve_fit
#create an array of values

x= np.linspace(0,1,21)

y = 2*x

#create a linear function    
    
def line_function(x, A, B):   #where A and B are parameters
    return (A*x)+B


#PURE DATA 

#use curve fit function
popt, pcov = curve_fit(line_function, x, y)


#Find the expected values
A_expected = popt[0]  
B_expected = popt[1]

#create a fitted line using predicted function
y_expected1 = line_function(x, A_expected, B_expected)


#plot the data and the fit function
plt.scatter(x, y, label = "data")
plt.plot(x, y_expected1, color = "orange", label = "fit")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Pure Data")
plt.legend()
```

__Ordinary Differential Equations__

```python
#By Hand 

# Define a function that computes the derivatives  
def derivs(L,g):
    # Define some variables for the air resistance
    
    k=6.0 #m
    m = 80.0 # kg
    L_Normal=30
    
    # derivative of height is velocity
    dldt = v
    
    # derivative of velocity is acceleration (gravity in this e.g.)
    dvdt = g + (-k*(L-L_Normal))/m
    
    return dldt, dvdt

def derivs_w_ar(L,g):
    k=6.0 #m
    m = 80.0 # kg
    A= 0.1 #m**2
    L_Normal=30
    # derivative of height is velocity
    dldt = v
    
    # derivative of velocity is acceleration (gravity in this e.g.)
    dvdt = g + ((-0.65 * A * v * abs(v))/m) +(-k*(L-L_Normal))/m
    
    return dldt, dvdt
```

```python
# Computing OED's Using Using Sci Py
#Without Air Resistance 
import matplotlib.pyplot as plt
%matplotlib inline

# Initialize variables
L = 0 # initial height; m
v = 0 # initial velocity; m/s 
g = 9.81 # gravity; m/s^2
t = 0 # initial time
tmax = 60 # Falling time
dt = 0.1 # timestep

# Initialize lists for storing data
Length = []
velocity = []
time = []
acceleration = []

# Append initial values to lists
Length.append(L)
velocity.append(v)
time.append(t)
acceleration.append(g)

# Create a time loop that will update the skydiver over time
# Use a while loop that will loop until t > tmax
while t <= tmax:
    
    # Compute derivatives for use in update equations
    dldt, dvdt = derivs_w_ar(L,g)
    
    # Update Equations
    L_new = L + dldt*dt # new height
    
    v_new = v + dvdt*dt # new velocity
    
    # Append values to height and velocity lists
    Length.append(L_new)
    velocity.append(v_new)
    acceleration.append(dvdt)
    
    # Update old height/velocity with new height
    L = L_new
    v = v_new
    
    # Increase time
    t += dt # t = t + dt
    
    # Update time list
    time.append(t)
    
# Plotting height/velocity/acceleration vs time 
plt.figure(1)
plt.plot(time, Length, color = 'green')
plt.xlabel('Time [seconds]')
plt.ylabel('Length [meters]')
plt.grid()

plt.figure(2)
plt.plot(time, velocity, color = 'purple')
plt.xlabel('Time [seconds]')
plt.ylabel('Veloctiy [meters/second]')
plt.grid()

plt.figure(3)
plt.plot(time, acceleration, color = 'orange')
plt.xlabel('Time [seconds]')
plt.ylabel('Acceleration [meters/second^2]')
plt.grid()
```
