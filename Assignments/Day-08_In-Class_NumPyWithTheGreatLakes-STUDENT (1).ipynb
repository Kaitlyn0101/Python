### **2.1 Means and Medians**

**What is the mean of a data set?**

The **mean**, also referred to as the average, is calculated by adding up all of the observations in a data set, and dividing by the number of observations in the data set.

$mean = \frac{\sum\limits_{i=1}^{N} x_i}{N}$, $N$ = number of observations, $x_i$ = an observation

More simply put,

mean = sum of observations / number of observations

The mean of a data set is useful because it provides a single number to describe a dataset that can be very large. However, the mean is sensitive to outliers (observations that are far from the mean), so it is best suited for data sets where the observations are close together.

**What is the median of a data set?**

The **median** of a data set is the middle value of a data set, or the value that divides the data set into two halves. The median also requires the data to be sorted from least to greatest. If the number of observations is odd, then the median is the middle value of the data set. If the number of observations is even, then the median is the average of the two middle numbers.

Unlike the mean, because the median is the midpoint of a data set, it is not strongly affected by a small number of outliers. 

### **Using numpy to Calculate Mean and Median**

We can calculate the mean and median from scratch, but we are going to introduce one option for functions in Python that will do the calculations for you. Because the calculations of mean and median are so useful, many Python packages include a function to calculate them, but here we are going to use the calculations from numpy.

The [documentation](https://numpy.org/doc/stable/reference/generated/numpy.mean.html) for numpy mean shows additional options, but the basic use is: 

`np.mean(data)`

Similarly, the median is calculated by:

`np.median(data)`

### **2.2 Calculating Standard Deviation**

One way to describe the distribution of a data set is the **standard deviation**. The standard deviation is the square root of the **variance** of the data. The variance is a measure of how "spread out" the distribution of the data is. More specifically, it is the squared difference between a single observation in a data set from the mean. The standard deviation is often represented with the greek symbol sigma ($\sigma$).

Mathematically, the equation for standard deivation is:

### $$ \sigma = \sqrt{\frac{\sum\limits_{i=1}^{N} (x_{i}-\mu)^2}{N}} $$

where the symbols in this equation represent the following:

* $\sigma$: Standard Deviation
* $\mu$: Mean
* $N$: Number of observations
* $x_{i}$: the value of dataset at position $i$

Similarly to `np.mean()` and `np.median()`, we can calculate the standard deviation and percentiles with numpy as follows:

`np.std(data)`

**In the cell below, calculate the standard deviation of the water level for each of the four lakes using the numpy function.**

```python
# put your code here

std_mhu=np.std(mhu_level)
std_er=np.std(er_level)
std_on=np.std(on_level)
std_sup=np.std(sup_level)

print("The standard deviation of each lake is", std_mhu, std_er, std_on, std_sup)
```

### 2.4 Contextualizing extreme weather events
___
Now that we've examined the current water levels, let's try to see whether the **variability** in water levels has increased over time.

1. Are seasonal changes in water levels increasing? ($\Delta t = 0.5$ years)
2. Are year-to-year variations in lake levels getting larger? ($\Delta t = 1$ year)
3. What about lake level variations over a 5-year period? ($\Delta t = 5$ years)

To answer these we will compare lake levels at a time $t$ with levels at an eariler time $t - \Delta t$.  The time lag $\Delta t$ to use is given above for the three different questions.

**Note that since there are 12 data points per year in each array, a $\Delta t$ of 0.5 years corresponds to a time lag of 6 data points.**  Here is an example of computing the variation in Lake Michigan water levels over a 3-month time period:

```python
def get_variation(date, lake_levels, time_lag):
    # computes the variation in lake levels over the 
    # specified time lag
    
    # number of elements in lake_levels
    n = len(lake_levels)

    # computing the variation using array indexing
    # Note: this reduces the number of elements!
    var = lake_levels[time_lag:n] - lake_levels[0:-time_lag]
    
    # return the variation, along with a set of dates 
    # that has the same number of elements
    return date[time_lag:n], var
```

```python
# length of time lag
time_lag = 3

# computing the variation over a six-month time period
mhu_date3, mhu_var3 = get_variation(mhu_date, mhu_level, time_lag)

# plotting this variation
plt.plot(mhu_date3,mhu_var3)
plt.xlabel('Year')
plt.ylabel('3-month variation (m)')
```

```python
print(len(mhu_var3))  
print(len(mhu_level))
```

&#9989;&nbsp; Using the `len()` function, find out how many elements are in `mhu_var3`.  How does this compare to the number of elements in `mhu_level`?  Does this make sense?

Yes, because there is a time lag of 3 elements meaning that there should be 3 less elements in mhu_var3.  

&#9989;&nbsp; **Use the `get_variation` function above to calculate the 6-month, 1-year and 5-year variations for Lake Michigan and plot them as subplots** in the same figure (3 rows, 1 column).

```python

#6 month variance
time_lag = 6

# computing the variation over a six-month time period
mhu_date6, mhu_var6 = get_variation(mhu_date, mhu_level, time_lag)

# plotting this variation
plt.figure(1)
plt.plot(mhu_date6,mhu_var6)
plt.xlabel('Year')
plt.ylabel('6-month variation (m)')
plt.title('6 Month Variation')

# 12 Month Variance 
time_lag = 12

# computing the variation over a six-month time period
mhu_date12, mhu_var12 = get_variation(mhu_date, mhu_level, time_lag)

# plotting this variation
plt.figure(2)
plt.plot(mhu_date12,mhu_var12, c='g')
plt.xlabel('Year')
plt.ylabel('12-month variation (m)')
plt.title("'12 month variation'")


# 5 year variance
time_lag = 60

# computing the variation over a six-month time period
mhu_date60, mhu_var60 = get_variation(mhu_date, mhu_level, time_lag)

# plotting this variation
plt.figure(3)
plt.plot(mhu_date60,mhu_var60, c='r')
plt.xlabel('Year')
plt.ylabel('60-month variation (m)')
plt.title("'60 month variation'")
```

&#9989;&nbsp; **Are these variations high in comparison to historical averages?**  **When was the largest 6-month, 1-year and 5-year variation recorded?**

```python

max6=np.max(mhu_var6)
max12=np.max(mhu_var12)
max60=np.max(mhu_var60)


 
print("The Largest Recorded Variance for a 6th month period", max6, "12 month period", max12, "5 year period", max60)
```
